```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
# Tạo thư mục để lưu ảnh
if not os.path.exists('images'):
    os.makedirs('images')
```

```python
# URL của trang cần crawl
url = 'https://hasaki.vn/catalogsearch/result/?q=s%C6%A1n%20m%C3%B3ng%20tay'
```

```python
# Gửi yêu cầu GET đến trang web
response = requests.get(url)
```

```python
# Kiểm tra xem yêu cầu thành công hay không
i = 0
if response.status_code == 200:
    # Phân tích HTML của trang web
    soup = BeautifulSoup(response.content, 'html.parser')

    # Tìm tất cả các thẻ hình ảnh trong trang
    images = soup.find_all('img',class_="img_thumb lazy")

    # Lặp qua từng hình ảnh và lưu xuống ổ đĩa
    for image in images:
        # Lấy URL của hình ảnh
        
        image_url = image['data-src']

        # Lấy tên tệp tin từ URL
        image_filename = os.path.basename(image_url)

        # Tạo URL tuyệt đối của hình ảnh
        absolute_image_url = url + image_url

        # Gửi yêu cầu GET đến URL của hình ảnh
        image_response = requests.get(absolute_image_url)

        # Kiểm tra xem yêu cầu thành công hay không
        if image_response.status_code == 200:
            # Lưu hình ảnh xuống thư mục 'images'
            with open('images/' + image_filename, 'wb') as f:
                f.write(image_response.content)

            print(f"Lưu ảnh {image_filename} thành công.")
            i+=1
        else:
            print(f"Lưu ảnh {image_filename} thất bại.")
else:
    print("Yêu cầu không thành công.")
    
print('số ảnh', i)
```

```python
base_url = 'https://abbeautyworld.com/search?type=product&q=filter%3D((title%3Aproduct%20contains%20s%C6%A1n%20m%C3%B3ng%20tay)%7C%7C(sku%3Aproduct%20contains%20s%C6%A1n%20m%C3%B3ng%20tay))&sortby=(quantity%3Aproduct%3Ddesc)&page='
```

```python
# Lặp qua các trang và cào hình ảnh
page = 1
i = 0
while True:
    # Tạo URL của trang hiện tại
    url = base_url + str(page)

    # Gửi yêu cầu GET đến URL
    response = requests.get(url)

    # Kiểm tra xem yêu cầu thành công hay không
    if response.status_code == 200:
        # Phân tích HTML của trang
        soup = BeautifulSoup(response.content, 'html.parser')

        # Tìm tất cả các thẻ hình ảnh
        image_tags = soup.find_all('img', class_="img-loop lazy")
        if not image_tags:
            break
        # Lặp qua danh sách các thẻ hình ảnh và lưu xuống ổ đĩa
        for image in image_tags:
            if 'sơn móng tay' in image.get('alt', '').lower():
                image_url = image['src']
                image_filename = os.path.basename(image_url)

                # Tạo URL tuyệt đối của hình ảnh
                absolute_image_url = 'http://' + image_url.lstrip('/')

                # Gửi yêu cầu GET đến URL của hình ảnh
                response = requests.get(absolute_image_url)

                if response.status_code == 200:
                    with open(f'images/{image_filename}', 'wb') as f:
                        f.write(response.content)

                    print(f"Lưu ảnh {image_filename} thành công.")
                    i += 1
                    
        page += 1

        # Kiểm tra nếu không tìm thấy thẻ chứa số trang hiện tại
        # if not soup.find('span', class_='pagination_current_page'):
        #     break

    else:
        print(f"Không thể truy cập trang {url}.")
        break

print(f"Tổng số ảnh sơn móng tay: {i}")
print('page', page)
```
